<h4>This dataset contains 11 columns, they are:</h4>


* <b> work_year</b>: The year the salary was paid.
* <b> experience_level</b>: The experience level when the salar was paid.
* <b> employment_type</b>:The employment type.
* <b> job_title</b>: The title of the role.
* <b> salary</b>: The total gross salary amount paid.
* <b> salary_currency</b>: The currency of the salary paid (in ISO 4217 currency code).
* <b> salary_in_usd</b>: The standardized salary paid (in USD).
* <b> employee_residence</b>: Employee's primary country of residence (in ISO 3166 country code).
* <b> remote_ratio</b>: The overall amount of work done remotely.
* <b> company_location</b>: The country of the employer's main office (in ISO 3166 country code).
* <b> company_size</b>: The size of the company in terms of the number of employees.
